## Example of price/volatility data

```python
fn = "european_wholesale_electricity_price_data_hourly/France.csv"

# parse dates, prefer local datetime column for daily aggregation
df = pd.read_csv(fn, parse_dates=['Datetime (Local)'])

# columns: Country, ISO3 Code, Datetime (UTC), Datetime (Local), Price (EUR/MWhe)
# keep only the local datetime and price
df = df[['Datetime (Local)', 'Price (EUR/MWhe)']].copy()
df = df.rename(columns={'Datetime (Local)': 'datetime', 'Price (EUR/MWhe)': 'price'})
df = df.sort_values('datetime').set_index('datetime')
df.index = pd.to_datetime(df.index)  # ensure datetime index

# ---------------------------
# Hourly returns
# ---------------------------
# compute log returns (natural log)
df['log_price'] = np.log(df['price'].replace(0, np.nan))
df['hourly_ret'] = df['log_price'].diff()
df = df.dropna(subset=['hourly_ret'])  # drop first NA

# ---------------------------
# Daily volatility measures
# ---------------------------
# Option A: Realized volatility (square-root of sum of squared intraday returns)
# Option B: Daily std of hourly returns scaled by sqrt(24) (if hours per day ~24)
# Guard against days with missing hours (e.g., DST or missing data)
min_hours_required = 20  # drop days with fewer than this many hourly observations

# attach a date column for grouping (local date)
df['date'] = df.index.normalize()  # midnight-normalized date

group = df.groupby('date')

# count hours per day
counts = group['hourly_ret'].count().rename('n_hours')

daily = pd.DataFrame({
    'n_hours': counts,
})

# realized volatility: sqrt(sum(r_i^2))
daily['realized_vol'] = group['hourly_ret'].apply(lambda x: np.sqrt(np.sum(x.values**2)))

# scaled std: std(hourly returns) * sqrt(24)  (common approximation)
daily['std_scaled'] = group['hourly_ret'].std() * np.sqrt(24)

# daily log return using last observation each day (close-to-close)
daily_close = df['log_price'].resample('D').last()
daily['daily_log_return'] = daily_close.diff().reindex(daily.index)

# filter days with too few hours (optional, recommended)
daily = daily[daily['n_hours'] >= min_hours_required]

# drop any rows with NaNs where it matters
daily = daily.dropna(subset=['realized_vol', 'std_scaled', 'daily_log_return'])

# ---------------------------
# Plotting
# ---------------------------
plt.rcParams['figure.figsize'] = (12, 4)

# Time series of realized volatility
plt.figure()
plt.plot(daily.index, daily['realized_vol'], label='Realized Vol (sqrt sum r^2)')
plt.plot(daily.index, daily['std_scaled'], label='std_scaled (std*sqrt(24))', alpha=0.7)
plt.legend()
plt.title('Daily Volatility Measures')
plt.xlabel('Date')
plt.ylabel('Volatility (log-return)')
plt.tight_layout()
plt.show()

# Histogram + QQplot of daily volatility
plt.figure()
plt.subplot(1,2,1)
plt.hist(daily['realized_vol'].dropna(), bins=40)
plt.title('Realized Daily Vol')
plt.subplot(1,2,2)
qqplot(daily['realized_vol'].dropna(), line='s', ax=plt.gca())
plt.title('QQ-plot: Realized Daily Vol')
plt.tight_layout()
plt.show()

# Time series of daily returns 
plt.figure()
plt.plot(daily.index, daily['daily_log_return'])
plt.title('Daily Log Returns (close-to-close)')
plt.axhline(0, color='k', linewidth=0.6)
plt.tight_layout()
plt.show()

# ACF of returns and squared returns
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plot_acf(daily['daily_log_return'].dropna(), lags=30, ax=plt.gca(), title='ACF: Daily Returns')
plt.subplot(1,2,2)
plot_acf(daily['daily_log_return'].dropna()**2, lags=30, ax=plt.gca(), title='ACF: Squared Daily Returns')
plt.tight_layout()
plt.show()

# Rolling volatility plot 
plt.figure()
rolling = daily['realized_vol'].rolling(window=30).mean()
plt.plot(daily.index, daily['realized_vol'], alpha=0.3, label='Daily realized vol')
plt.plot(daily.index, rolling, label='30-day rolling mean', linewidth=2)
plt.legend()
plt.title('Realized Volatility + 30-day Rolling Mean')
plt.tight_layout()
plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.regression.linear_model import OLS

all_preds = all_preds.sort_values('DATE').reset_index(drop=True)

for c in ['DE', 'FR']:
    idx = all_preds[all_preds['COUNTRY'] == c].index  # integer positions in all_preds
    sub = all_preds.loc[idx].copy()

    err = (sub['pred'] - sub['true']).reset_index(drop=True)

    # Build AR(7) design matrix on the error series
    X_ar = pd.concat([err.shift(i) for i in range(1, 8)], axis=1)
    X_ar.columns = [f'lag_{i}' for i in range(1, 8)]

    valid = X_ar.dropna().index  # rows where all 7 lags are available
    X_fit = sm.add_constant(X_ar.loc[valid])
    y_fit = err.loc[valid]

    ar_model = OLS(y_fit, X_fit).fit()
    correction = ar_model.fittedvalues  # indexed by valid (0-based within sub)

    # Map back: valid is 0-based within sub, idx gives all_preds positions
    all_preds.loc[idx[valid], 'pred'] -= correction.values
```

```python
"""
Notional conversion: dimensionless log-vol residual P&L → EUR

ECONOMIC INTERPRETATION
-----------------------
Your position is in log-volatility space:
  pnl_dimensionless = position × (log_rv_actual - log_har_forecast)

To convert to EUR, you need a notional that maps log-vol moves to price moves.
The standard approach for power markets:

  EUR_pnl = position × vega × price_level × notional_MW × days_per_year^0.5

where vega is approximately 1.0 for a variance swap (the natural instrument
for this strategy), and price_level is the day-ahead baseload price in EUR/MWh.

INSTRUMENT CHOICES
------------------
1. Variance swap (most direct): pays realized_var - strike_var
   Notional in EUR/vol² per MWh. Your log-vol residual maps directly.

2. Straddle on baseload futures (practical approximation):
   delta-neutral options position. Vega ≈ price × vol × sqrt(T).
   When you buy a straddle, you profit from vol being higher than implied.

3. Baseload futures with dynamic sizing (most liquid, indirect):
   Size your long/short baseload position proportional to predicted vol.
   Higher predicted vol → larger position (you expect bigger moves).
   P&L comes from directional price moves, not vol directly.

We implement option 3 as it uses instruments you definitely have prices for.
The position sizing rule: scale notional_MW by your predicted vol signal.

NOTIONAL DEFINITION
-------------------
1 baseload lot on EEX = 1 MW for all hours of the delivery period.
For a 1-day baseload contract: 1 lot × 24 hours = 24 MWh.
At a price of P EUR/MWh, 1 lot has a notional value of 24 × P EUR.

A typical institutional position: 10-50 MW per country.
We use notional_mw = 10 MW as a conservative base case.
"""

import numpy as np
import pandas as pd


def convert_pnl_to_eur(
    bt: pd.DataFrame,
    panel: pd.DataFrame,               # your full panel with DATE column
    price_col_de: str = 'DE_price',    # daily mean price EUR/MWh in panel
    price_col_fr: str = 'FR_price',    # daily mean price EUR/MWh in panel
    notional_mw: float = 10.0,         # MW per country per trade
    hours_per_day: int = 24,
    vol_scale_factor: float = None,    # if None, computed from data
    verbose: bool = True,
) -> pd.DataFrame:
    """
    Convert dimensionless P&L to EUR-denominated P&L.

    METHOD
    ------
    The dimensionless P&L is:
        pnl_dim = position × (log_rv_actual - log_har_forecast)

    The log-vol residual has units of log(EUR/MWh volatility). To convert:

    Step 1: Map log-vol residual to realised price return
        The relationship between log-vol residual and price return is:
        price_return ≈ vol_level × log_vol_residual
        where vol_level = rolling mean of realized_vol (price vol in EUR/MWh)

    Step 2: Multiply by notional
        EUR_pnl = position × vol_level × price_level × notional_MW × 24

    This is the P&L from a delta-hedged options position (long/short gamma)
    sized at notional_mw MW.

    ALTERNATIVELY (simpler, used here):
        Scale the dimensionless P&L by a constant factor:
        scale = median(price_level) × median(vol_level) × notional_MW × 24
        EUR_pnl = pnl_dim × scale

    The constant-scale approach is appropriate for backtesting because it
    preserves the Sharpe ratio and relative performance metrics exactly.
    Only the absolute P&L numbers change.
    """
    df = bt.copy()

    # ---- Get price data from panel ----
    # Panel may have one row per country or one row per date
    # Try both structures
    panel_daily = panel.copy()
    if 'DATE' in panel_daily.columns:
        panel_daily = panel_daily.set_index('DATE')
    panel_daily.index = pd.to_datetime(panel_daily.index)

    # Extract DE and FR prices
    # Your panel has prices embedded in realized_vol computation
    # The EMBER price data was used: price_col should be the daily mean price
    price_de = None
    price_fr = None

    if price_col_de in panel_daily.columns:
        price_de = panel_daily[price_col_de]
    if price_col_fr in panel_daily.columns:
        price_fr = panel_daily[price_col_fr]

    # If price columns not in panel, try to compute from realized_vol
    # using the relationship: realized_vol ≈ price × intraday_vol_fraction
    # Fallback: use a fixed representative price
    if price_de is None or price_fr is None:
        if verbose:
            print("WARNING: Price columns not found in panel. "
                  "Using fallback price estimates.")
            print(f"  Looking for: '{price_col_de}', '{price_col_fr}'")
            print(f"  Available columns with 'price': "
                  f"{[c for c in panel_daily.columns if 'price' in c.lower()]}")
            print("  Falling back to representative prices: DE=80, FR=80 EUR/MWh")
            print("  Re-run after adding price columns to panel for accurate EUR P&L.")

        # Representative prices (rough European average 2015-2026)
        # You should replace these with actual prices from your EMBER data
        date_index = pd.to_datetime(df['DATE'].unique())
        price_de = pd.Series(80.0, index=date_index, name=price_col_de)
        price_fr = pd.Series(80.0, index=date_index, name=price_col_fr)

    # ---- Get vol levels from panel ----
    rv_de = panel_daily.get('DE_realized_vol', None)
    rv_fr = panel_daily.get('FR_realized_vol', None)

    # ---- Merge prices and vol into bt ----
    price_map_de = price_de.to_dict()
    price_map_fr = price_fr.to_dict()

    df['DATE_dt'] = pd.to_datetime(df['DATE'])
    df['price_level'] = df.apply(
        lambda r: price_map_de.get(r['DATE_dt'], 80.0)
        if r['COUNTRY'] == 'DE'
        else price_map_fr.get(r['DATE_dt'], 80.0),
        axis=1,
    )

    if rv_de is not None and rv_fr is not None:
        rv_map_de = rv_de.to_dict()
        rv_map_fr = rv_fr.to_dict()
        df['vol_level'] = df.apply(
            lambda r: rv_map_de.get(r['DATE_dt'], rv_de.median())
            if r['COUNTRY'] == 'DE'
            else rv_map_fr.get(r['DATE_dt'], rv_fr.median()),
            axis=1,
        )
    else:
        # Fallback: infer vol level from realized_vol in bt if present
        if 'DE_realized_vol' in df.columns:
            df['vol_level'] = df.apply(
                lambda r: r.get('DE_realized_vol', 0.02)
                if r['COUNTRY'] == 'DE'
                else r.get('FR_realized_vol', 0.02),
                axis=1,
            )
        else:
            df['vol_level'] = 0.02   # ~2% daily vol, typical for power

    # ---- Compute scaling factor ----
    # EUR_pnl = pnl_dimensionless × price_level × vol_level × notional_MW × hours
    #
    # Intuition:
    #   pnl_dimensionless is in units of "log-vol points"
    #   × vol_level converts to "price return fraction" (approx)
    #   × price_level converts to EUR/MWh
    #   × notional_mw × hours converts to EUR per trade
    #
    # vol_scale_factor overrides the per-row vol_level if set
    # (use this for a fixed scale that gives stable Sharpe comparison)

    if vol_scale_factor is not None:
        df['eur_scale'] = vol_scale_factor * notional_mw * hours_per_day
    else:
        df['eur_scale'] = (
            df['vol_level']
            * df['price_level']
            * notional_mw
            * hours_per_day
        )

    # ---- Convert P&L ----
    df['pnl_eur']             = df['pnl']             * df['eur_scale']
    df['pnl_before_cost_eur'] = df['pnl_before_cost'] * df['eur_scale']
    df['tcost_eur']           = df['tcost']            * df['eur_scale']

    df.drop(columns=['DATE_dt'], inplace=True)

    if verbose:
        daily_pnl_eur = df.groupby('DATE')['pnl_eur'].sum()
        total_eur = daily_pnl_eur.sum()
        ann_eur   = daily_pnl_eur.mean() * 252
        sharpe    = (daily_pnl_eur.mean() / (daily_pnl_eur.std() + 1e-9)) * np.sqrt(252)
        med_scale = df['eur_scale'].median()

        print(f"\n=== EUR P&L Summary ===")
        print(f"Notional:           {notional_mw:.0f} MW per country "
              f"= {notional_mw * hours_per_day * df['price_level'].median():.0f} EUR/day notional")
        print(f"Median scale factor:{med_scale:.2f} EUR per unit P&L")
        print(f"Total P&L:          {total_eur:,.0f} EUR  "
              f"({total_eur/1e6:.2f}M EUR)")
        print(f"Annualised P&L:     {ann_eur:,.0f} EUR/year")
        print(f"Sharpe (EUR):       {sharpe:.4f}  "
              f"(should match dimensionless: scaling preserves Sharpe)")
        print(f"Total Cost (EUR):   {df['tcost_eur'].sum():,.0f} EUR")
        print(f"Max daily loss:     {daily_pnl_eur.min():,.0f} EUR")
        print(f"Max daily gain:     {daily_pnl_eur.max():,.0f} EUR")

        # Annualised return on notional
        # Notional = 2 × notional_mw × hours × median_price (long one, short one)
        total_notional = 2 * notional_mw * hours_per_day * df['price_level'].median()
        ann_return_pct = ann_eur / total_notional * 100
        print(f"Total notional:     {total_notional:,.0f} EUR")
        print(f"Ann. return on notional: {ann_return_pct:.1f}%")

    return df


def build_price_panel(
    price_file_de: str,
    price_file_fr: str,
) -> pd.DataFrame:
    """
    Build a daily price panel from your EMBER CSV files.
    Returns dataframe with DATE, DE_price, FR_price.

    Matches your existing price loading code:
        prices = pd.read_csv(price_file, parse_dates=['Datetime (Local)'])
        prices.columns = ['datetime', 'price']
    """
    def _load(path, col_name):
        p = pd.read_csv(path, parse_dates=['Datetime (Local)'])
        p = p[['Datetime (Local)', 'Price (EUR/MWhe)']].copy()
        p.columns = ['datetime', 'price']
        p = p.sort_values('datetime').set_index('datetime')
        p['price'] = p['price'].replace(0, np.nan)
        # Daily mean, median, and vol
        daily = p['price'].resample('D').agg(
            mean='mean',
            median='median',
            std='std',
            min='min',
            max='max',
        )
        daily.columns = [f'{col_name}_{s}' for s in daily.columns]
        daily[f'{col_name}'] = daily[f'{col_name}_mean']  # convenience alias
        daily.index = daily.index.normalize()
        return daily

    de = _load(price_file_de, 'DE_price')
    fr = _load(price_file_fr, 'FR_price')

    combined = de.join(fr, how='outer')
    combined.index.name = 'DATE'
    combined = combined.reset_index()

    print(f"Price panel: {len(combined)} days, "
          f"DE mean={combined['DE_price'].mean():.1f} EUR/MWh, "
          f"FR mean={combined['FR_price'].mean():.1f} EUR/MWh")
    return combined


def sensitivity_to_notional(
    bt: pd.DataFrame,
    panel: pd.DataFrame,
    notional_range: list = None,
    **kwargs,
) -> pd.DataFrame:
    """
    Show how EUR P&L scales with notional MW assumption.
    Sharpe is invariant to notional — only absolute P&L changes.
    """
    if notional_range is None:
        notional_range = [1, 5, 10, 25, 50, 100]

    rows = []
    for mw in notional_range:
        bt_eur = convert_pnl_to_eur(
            bt, panel, notional_mw=mw, verbose=False, **kwargs
        )
        daily = bt_eur.groupby('DATE')['pnl_eur'].sum()
        rows.append({
            'notional_MW':    mw,
            'total_pnl_eur':  daily.sum(),
            'ann_pnl_eur':    daily.mean() * 252,
            'sharpe':         (daily.mean() / (daily.std() + 1e-9)) * np.sqrt(252),
            'max_loss_eur':   daily.min(),
            'max_gain_eur':   daily.max(),
        })

    result = pd.DataFrame(rows)
    print("\n=== Notional Sensitivity ===")
    print(result.to_string(index=False, float_format='{:,.0f}'.format))
    return result
```

```python
def convert_pnl_to_eur(
    bt: pd.DataFrame,
    price_panel: pd.DataFrame,         # output of build_price_panel()
    notional_mw: float = 10.0,
    hours_per_day: int = 24,
    verbose: bool = True,
) -> pd.DataFrame:
    """
    Convert dimensionless P&L to EUR.
    
    Interpretation: pnl_dimensionless is treated as a fractional return
    on a notional position of notional_mw MW at the prevailing price.
    
    EUR_pnl_t = pnl_dim_t × notional_mw × hours_per_day × price_t
    
    This is consistent with trading baseload futures:
    1 MW baseload = 24 MWh/day = 24 × price EUR/day notional.
    """
    df = bt.copy()
    
    # Merge daily prices
    price_panel = price_panel.copy()
    price_panel['DATE'] = pd.to_datetime(price_panel['DATE'])
    df['DATE'] = pd.to_datetime(df['DATE'])
    
    df = df.merge(
        price_panel[['DATE', 'DE_price', 'FR_price']],
        on='DATE', how='left'
    )
    
    # Price for this row's country
    df['price_t'] = np.where(
        df['COUNTRY'] == 'DE',
        df['DE_price'].fillna(df['DE_price'].median()),
        df['FR_price'].fillna(df['FR_price'].median()),
    )
    
    # Scale: MW × hours × EUR/MWh = EUR per unit of dimensionless P&L
    df['eur_scale'] = notional_mw * hours_per_day * df['price_t']
    
    df['pnl_eur']             = df['pnl']             * df['eur_scale']
    df['pnl_before_cost_eur'] = df['pnl_before_cost'] * df['eur_scale']
    df['tcost_eur']           = df['tcost']           * df['eur_scale']
    
    if verbose:
        daily_eur = df.groupby('DATE')['pnl_eur'].sum()
        total_eur = daily_eur.sum()
        ann_eur   = daily_eur.mean() * 252
        sharpe    = (daily_eur.mean() / (daily_eur.std() + 1e-9)) * np.sqrt(252)
        
        avg_price  = (df['DE_price'].median() + df['FR_price'].median()) / 2
        gross_notional = 2 * notional_mw * hours_per_day * avg_price
        ann_return_pct = ann_eur / gross_notional * 100
        
        print(f"\n=== EUR P&L ({notional_mw:.0f} MW notional) ===")
        print(f"Avg price:           DE={df['DE_price'].median():.1f}  "
              f"FR={df['FR_price'].median():.1f}  EUR/MWh")
        print(f"Gross notional:      {gross_notional:,.0f} EUR/day  "
              f"(long {notional_mw:.0f}MW + short {notional_mw:.0f}MW × 24h)")
        print(f"Total P&L:           {total_eur:,.0f} EUR  "
              f"({total_eur/1e6:.2f}M EUR over {len(daily_eur)} days)")
        print(f"Annualised P&L:      {ann_eur:,.0f} EUR/year")
        print(f"Ann. return on notional: {ann_return_pct:.1f}%")
        print(f"Sharpe (EUR):        {sharpe:.4f}  "
              f"← should match dimensionless {1.93:.2f} closely")
        print(f"Max daily loss:      {daily_eur.min():,.0f} EUR")
        print(f"Max daily gain:      {daily_eur.max():,.0f} EUR")
        print(f"Total costs:         {df['tcost_eur'].sum():,.0f} EUR")
    
    return df
```
